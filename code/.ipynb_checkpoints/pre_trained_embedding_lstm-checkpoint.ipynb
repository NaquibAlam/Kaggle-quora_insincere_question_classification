```python
def make_wordcloud(text):
    stopwords = set(STOPWORDS)
    wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=40, 
                          random_state=42
                         ).generate(text)
    print(wordcloud)
    fig = plt.figure(figsize= (20,16))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    
make_wordcloud(train.loc[train['target']==1,'question_text'].str.cat(sep= ' '))   
```

```python
def make_vocab_cnt(sentences):
#     return Counter(texts.split(' '))
    vocab= {}
    for sentence in sentences:
        for word in sentence:
            if word in vocab:
                vocab[word]+=1
            else:
                vocab[word]= 1
    return vocab
#     retutn vocab_cnt

# vocab_cnt= make_vocab_cnt(train['question_text'].str.cat(sep=' '))
vocab_cnt= make_vocab_cnt(train['question_text'].apply(lambda x: x.split()).values)
vocab_low_cnt= make_vocab_cnt(train['question_text'].apply(lambda x:x.lower()).apply(lambda x: x.split()).values)
# vocab_low_cnt= make_vocab_cnt(train['question_text'].str.cat(sep=' ').lower())
print(len(list(vocab_cnt.keys())), len(list(vocab_low_cnt.keys())))
```

```python
def add_lower(vocab_cnt, embedding):
    count= 0
    for word in list(vocab_cnt.keys()):
        if word in embedding and word.lower() not in embedding:
            embedding[word.lower()]= embedding[word]
            count+=1
    print(f'added {count} words to embdding')
print("BEFORE: ", len(glove_dict))
add_lower(vocab_cnt, glove_dict)
print("AFTER: ", len(glove_dict))
print("BEFORE: ", len(wiki_dict))
add_lower(vocab_cnt, wiki_dict)
print("AFTER: ", len(wiki_dict))
print("BEFORE: ", len(google_dict))
add_lower(vocab_cnt, google_dict)
print("AFTER: ", len(google_dict))
```

```python
def calc_coverage(vocab, embedding):
    vocab_words= set(vocab.keys())
    embed_words= set(embedding.keys())
    common_words= vocab_words.intersection(embed_words)
    total_words_match= 0
    for word in common_words:
        total_words_match+=vocab[word]
    oov_words= vocab_words - common_words
    oov_words_cnt= []
    for word in oov_words:
        oov_words_cnt.append((word,vocab[word]))
    oov_words_cnt.sort(key= itemgetter(1), reverse= True)
    total_corpus_len= sum(list(vocab.values()))
    pcntg_common_words= (len(common_words)/len(vocab_words))*100
    total_pcntg_words_match= (total_words_match/total_corpus_len)*100
    print(len(vocab_words), len(embed_words), len(common_words), pcntg_common_words, total_pcntg_words_match)
    return oov_words_cnt
oov_words= calc_coverage(vocab_cnt, glove_dict)
oov_words_low= calc_coverage(vocab_low_cnt, glove_dict)
print(oov_words[:20])
print()
print(oov_words_low[:20])
```

```python
def correct_contractions(text, mapping):
    specials = ["’", "‘", "´", "`"]
    for s in specials:
        text = text.replace(s, "'")
    text = ' '.join([mapping[t] if t in mapping else t for t in text.split(" ")])
    return text
train['corrected_text']= train['question_text'].apply(lambda x:x.lower()).apply(lambda x:correct_contractions(x, contraction_mapping))
# vocab_cnt_corrected= make_vocab_cnt(train['corrected_text'].str.cat(sep=' '))
vocab_cnt_corrected= make_vocab_cnt(train['corrected_text'].apply(lambda x: x.split()).values)
# del vocab_cnt_corrected['']
oov_words_corrected= calc_coverage(vocab_cnt_corrected, glove_dict)
print(oov_words_corrected[:20])
```

```python
def clean_numbers(x):

    x = re.sub('[0-9]{5,}', '#####', x)
    x = re.sub('[0-9]{4}', '####', x)
    x = re.sub('[0-9]{3}', '###', x)
    x = re.sub('[0-9]{2}', '##', x)
    return x
```

```python
def find_frequent_words(sentences, stop_words, top_words=100):
    vocab_cnt= {}
    for sentence in sentences:
        for word in sentence:
            if word in stop_words: continue
            if word in vocab_cnt:
                vocab_cnt[word]+=1
            else:
                vocab_cnt[word]=1
    vocab_cnt_sorted= sorted(vocab_cnt.items(), key= itemgetter(1), reverse= True)
    return vocab_cnt_sorted[:100]

top_k_pos= find_frequent_words(train.loc[train['target']==1, 'corrected_text'].apply(lambda x:x.split()).values, stop_words)
top_k_pos
```
